# Ecommerce Data Exploration and Cleaning

This notebook demonstrates the basics of loading a CSV, exploring it, handling missing values, filtering and selecting columns, removing duplicates, creating a derived column, and saving a cleaned CSV.

If `ecommerce_dataset.csv` is present in the same folder, the notebook uses it. Otherwise, it falls back to a small demo ecommerce dataframe so the workflow still runs end to end.

```python
import pandas as pd
from pathlib import Path

input_path = Path("combined_dataset.csv")
output_path = Path("combined_dataset_cleaned.csv")

demo_data = pd.DataFrame([
    {"product_id": 1, "title": "T-shirt", "price": 20, "quantity": 2, "category": "Clothing", "rating": 4.3},
    {"product_id": 2, "title": "Sneakers", "price": 80, "quantity": 1, "category": "Footwear", "rating": 4.7},
    {"product_id": 3, "title": "Headphones", "price": 50, "quantity": 3, "category": "Electronics", "rating": None},
    {"product_id": 4, "title": "Mug", "price": None, "quantity": 4, "category": "Home", "rating": 4.1},
    {"product_id": 1, "title": "T-shirt", "price": 20, "quantity": 2, "category": "Clothing", "rating": 4.3},
    {"product_id": 6, "title": "Jacket", "price": 120, "quantity": None, "category": "Clothing", "rating": 4.8},
    {"product_id": 7, "title": "Notebook", "price": 5, "quantity": 10, "category": "Stationery", "rating": 4.0},
    {"product_id": 8, "title": "Socks", "price": 10, "quantity": 5, "category": "Clothing", "rating": None},
])

df = pd.read_csv(input_path) if input_path.exists() else demo_data.copy()
source_name = input_path.name if input_path.exists() else "demo combined data"
print(f"Loaded {source_name} with shape: {df.shape}")
df.head()
```

```python
print("Tail:")
display(df.tail())
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("Data types:")
print(df.dtypes)
```

```python
print("Missing values per column:")
print(df.isna().sum())

for column in df.select_dtypes(include="object").columns:
    df[column] = df[column].fillna("Unknown")

for column in df.select_dtypes(include="number").columns:
    df[column] = df[column].fillna(df[column].median())

df = df.drop_duplicates().reset_index(drop=True)
df
```

```python
filtered_rows = df[df["category"].isin(["Clothing", "Electronics"])]
selected_columns = df[["product_id", "title", "price", "quantity", "category", "rating"]]

df["total_amount"] = df["price"] * df["quantity"]

display(filtered_rows)
display(selected_columns)
display(df)
```

```python
df.to_csv(output_path, index=False)
print(f"Cleaned dataset saved to: {output_path.resolve()}")
```

## Brief Summary
- Loaded an ecommerce dataset, or a demo fallback if the CSV was absent.
- Explored the data with head/tail, shape, columns, and data types.
- Filled missing values, removed duplicates, filtered rows, selected columns, and created `total_amount = price * quantity`.
- Saved the cleaned file as `ecommerce_dataset_cleaned.csv`.
